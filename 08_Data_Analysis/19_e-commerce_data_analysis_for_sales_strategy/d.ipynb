```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Sales_January_2019.csv')
df1['month'] = 'jan'

df2 = pd.read_csv('Sales_February_2019.csv')
df2['month'] = 'feb'

df3 = pd.read_csv('Sales_March_2019.csv')
df3['month'] = 'mar'

df4 = pd.read_csv('Sales_April_2019.csv')
df4['month'] = 'apr'

df5 = pd.read_csv('Sales_May_2019.csv')
df5['month'] = 'may'

df6 = pd.read_csv('Sales_June_2019.csv')
df6['month'] = 'jun'

df7 = pd.read_csv('Sales_July_2019.csv')
df7['month'] = 'jul'

df8 = pd.read_csv('Sales_August_2019.csv')
df8['month'] = 'aug'

df9 = pd.read_csv('Sales_September_2019.csv')
df9['month'] = 'seb'

df10 = pd.read_csv('Sales_October_2019.csv')
df10['month'] = 'oct'

df11 = pd.read_csv('Sales_November_2019.csv')
df11['month'] = 'nov'

df12 = pd.read_csv('Sales_December_2019.csv')
df12['month'] = 'dec'
```

```python
all_df = all_df.dropna()
all_df.isnull().sum()
```

```python
all_df['quantity_ordered'].unique()
```

```python
all_df['price_each'].unique()
```

```python
df = all_df[all_df.quantity_ordered != 'Quantity Ordered']
df['quantity_ordered'].unique()
```

```python
df = all_df[all_df.price_each != 'Price Each']
df['price_each'].unique()
```

```python
df = df.astype({'quantity_ordered': int , 'price_each':float})
df.info()
```

```python
df['price_each'].unique()
```

```python
df.price_each.sum()
```

```python
df['sales'] = df.quantity_ordered * df.price_each
df
```

```python
plt.subplots(figsize=(16,10))
ax=sns.barplot(x='month' , y='sales' ,data=df)
for c in ax.containers:
    ax.bar_label(c)
```

```python
d = df.groupby('month').sales.sum().sort_values()
d.head(12)
```

```python
month_sales = df.groupby('month').sales.agg(sales_sum = 'sum').sort_values(by='sales_sum' , ascending=False).reset_index()
month_sales
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax = sns.barplot(x='sales_sum' , y='month' , data= month_sales )
plt.title('month VS sale')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
def get_city(address):
    return address.split(',')[1].strip(' ')

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['city'] = df.purchase_address.apply(lambda x: get_city(x) + '(' + get_state(x) + ')')
df.head()
```

```python
city_sales = df.groupby('city').sales.agg(city_sale = 'sum').sort_values(by='city_sale' , ascending=False).reset_index()
city_sales
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax = sns.barplot(x='city_sale' , y='city' , data= city_sales )
plt.title('city VS sale')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
pro_qun = df.groupby('product').quantity_ordered.agg(pro_qun = 'sum').sort_values(by='pro_qun' , ascending=False).reset_index()
pro_qun
```
