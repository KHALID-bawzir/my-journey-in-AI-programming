```python
df.discount_rate_per_visited_products = df.discount_rate_per_visited_products.replace(',' , '.' , regex=True).astype(float)
df.product_detail_view_per_app_session = df.product_detail_view_per_app_session.replace(',' , '.' ,regex=True).astype(float)
df.add_to_cart_per_session = df.add_to_cart_per_session.replace(',' , '.' ,regex=True).astype(float)
df.avg_order_value = df.avg_order_value.replace(',' , '.' ,regex=True).astype(float)

df.info()
```

```python
columns_index =['account_length', 'add_to_wishlist', 'desktop_sessions', 'app_sessions', 'desktop_transactions',
'total_product_detail_views', 'session_duration', 'promotion_clicks', 'avg_order_value', 'sale_product_views',
'discount_rate_per_visited_products', 'product_detail_view_per_app_session', 'app_transactions',
'add_to_cart_per_session', 'customer_service_calls']
fig , axes = plt.subplots(5,3 , figsize=(30,30))
fig.suptitle("Distribution of numeric variables", size = 30)

for i in range(5):
    for j in range(3):
        sns.histplot(df[columns_index[3*i+j]] , ax=axes[i,j])
```

```python
df.churn.value_counts()
```

```python
fig, axes = plt.subplots(5, 3, figsize=(30,30))
fig.suptitle("Distribution of numeric variables", size=30)

axes = axes.flatten()

for col, ax in zip(columns_index, axes):
    sns.histplot(df[col], ax=ax)
    ax.set_title(col)


plt.show()
```

```python
fig , ax = plt.subplots(1,2 , figsize=(16,8))

plt.pie(df.churn.value_counts() , labels=['not churn' , 'churn'] ,colors=['green' , 'red'] ,autopct='%.0f%%' ,textprops={'fontsize':20} )

cp =sns.countplot(x = df.churn , ax=ax[0] , hue=df.churn)
cp.set_xlabel('Churn' , fontsize=15)
cp.set_ylabel('number of cas' , fontsize=15)
for container in cp.containers:
    cp.bar_label(container)
plt.show()
```

```python
categorical_columns = ["location_code", "push_status", "credit_card_info_save"]

# تصفية العملاء الذين قاموا بالإلغاء فقط
churn_df = df[df['churn'] == 0]  # تأكد أن عمود "churn" يمثل حالة الإلغاء

for i in churn_df.select_dtypes(include=['object']).columns.values:
    churn_df[i] = churn_df[i].replace(["yes", "no"], [1, 0])

fig, ax = plt.subplots(1, len(categorical_columns), figsize=(15, 6))
for i, a in zip(categorical_columns, ax):
    temp = churn_df[i].value_counts()
    a.set_title("Churn Counts Of " + str(i))
    sns.barplot(ax=a, x=temp.index, y=temp.values ,hue=temp.values)
plt.show()
```

```python
df_object=df.select_dtypes(include=['object']).columns.values
df_ss =['credit_card_info_save','location_code' ,'push_status']
for i in df_object:
    df[i]=df[i].replace(['yes' , 'no'] , [1,0])
fig,axes=plt.subplots(2,2 , figsize=(16,8))
axes=axes.flatten()
for i , a in zip(df_ss , axes):
    temp = df[i].value_counts()
    a.set_title('Churn count of ' + str(i))
    sns.barplot(ax = a , x = temp.index , y= temp.values , hue=temp.values)
```

```python
df_object = df.select_dtypes(include=['object']).columns

for col in df_object:
    df[col] = df[col].replace({'yes': 1, 'no': 0})

fig, axes = plt.subplots(2, len(df_ss), figsize=(16,6))
axes = axes.flatten()

for col, ax in zip(df_ss, axes):
    temp = df[col].value_counts()
    sns.barplot(x=temp.index, y=temp.values, ax=ax)
    ax.set_title(f'Churn count of {col}')

plt.tight_layout()
plt.show()
```
