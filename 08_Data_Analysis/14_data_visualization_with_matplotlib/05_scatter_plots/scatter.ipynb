```python
x = [5,7,8,7,2,12,2, 9]
y = [95,90,90,88,111,86,103,87]
colors =[5,13,20,43,60,75,95,100] 

plt.subplot(1,2,1)
plt.title(label= 'Y value')
plt.scatter(x,y ,c=y ,cmap='vanimo')
plt.colorbar()

plt.subplot(1,2,2)
plt.scatter(x,y ,c=colors ,cmap='vanimo')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('Students.csv' ,sep=';')
df.info()
```

```python
x = df['math score']
y = df['reading score']
z = df['writing score']
plt.subplot(1,2,1)
plt.scatter(x,y ,c=y ,cmap='vanimo')
plt.xlabel('math score')
plt.ylabel('reading score')
plt.colorbar()

plt.subplot(1,2,2)
plt.scatter(y,z ,c=y ,cmap='gist_ncar')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.colorbar()

plt.tight_layout()
plt.show()
```
