```python
import pandas as pd
df = pd.read_csv('car_malfunctions.csv')
df.info()
```

```python
import matplotlib.pyplot as plt
dt = df['DaysOff']
plt.plot(dt,
         marker = 'o' ,
         ls = ':' ,
         color = 'green' ,
         linewidth=1
        )

plt.title('DaysOff' , fontsize=33)
plt.xlabel('X' , fontsize=33)
plt.ylabel('y' , fontsize=33)

plt.show()
```

```python
y1=df['malfunctions']
y2=df['Age']
plt.plot(y1 ,marker = 'o' ,ls = ':' ,color = 'green' ,linewidth=1)
plt.plot(y2 ,marker = 'o' ,ls = ':' ,color = 'red' ,linewidth=1)
plt.show()
```
