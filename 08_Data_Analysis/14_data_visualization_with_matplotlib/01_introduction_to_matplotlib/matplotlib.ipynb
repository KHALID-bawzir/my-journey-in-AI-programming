```python
import matplotlib.pyplot as plt
```

```python
ydata = [2,3,1,4]
xdata = [6,5,3,9]
```

```python
plt.plot(ydata, linewidth=3)
plt.show()
```

```python
plt.plot(xdata , ydata , linewidth=3)
plt.show()
```

```python
plt.title('DATA' ,fontsize=44)
plt.xlabel('data x' ,fontsize=44 )
plt.ylabel('data y' ,fontsize=33 )

plt.show()
```
