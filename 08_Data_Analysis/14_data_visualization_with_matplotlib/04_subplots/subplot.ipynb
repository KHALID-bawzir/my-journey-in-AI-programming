```python
import matplotlib.pyplot as plt
y=[10,20,10,40]
```

```python
mylabels=['a' , 'b' , 'c' , 'd']
myexplod = [0.2,0.1,0,0]
mycolor=['#800080','#008080','#800000','#808080']


plt.subplot(1,3,1)
plt.pie(y ,  labels=mylabels  , explode=myexplod ,shadow=True , colors=mycolor )

plt.subplot(1,3,2)
plt.plot( y , linewidth=3)
plt.title('data' ,fontsize=44)
plt.ylabel('data y' ,fontsize=33 )

plt.subplot(1,3,3)
plt.plot(y ,marker = 'o' ,ls = ':' ,color = 'green' ,linewidth=1)

plt.tight_layout()
```
