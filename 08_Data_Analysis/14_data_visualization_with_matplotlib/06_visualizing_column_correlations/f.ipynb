```python
import pandas as pd
df = pd.read_csv('stdMarks.csv')
col = df[['Math' , 'Phy' , 'Lang']]
colc =col.corr(method='pearson')
colc
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cax = ax.matshow(colc , vmin = -1 , vmax = 1 ,cmap='PiYG')
fig.colorbar(cax)
tk = [0,1,2]
ax.set_xticks(tk)
ax.set_yticks(tk)
ax.set_xticklabels(['Math' , 'Phy' , 'Lang'])
ax.set_yticklabels(['Math' , 'Phy' , 'Lang'])
```
