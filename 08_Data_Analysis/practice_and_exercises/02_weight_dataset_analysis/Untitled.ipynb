```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hw_200.csv')
df.head(10)
```

```python
df.info()
```

```python
df.rename(columns={
    df.columns[2]: "weight_P",
    df.columns[1]: "height_in",
    df.columns[0]: "index"
}, inplace=True)
```

```python
df.columns
```

```python
df.weight.count()
```

```python
df['weight_kg'] = df['weight'] * 0.453592
df
```

```python
df['height_cm'] = df['height'] * 2.54
df.height_cm
```

```python
df.loc[100:110]
```

```python
fig , ax = plt.subplots(1,2 , figsize=(20,5))
sns.histplot(x='height_cm' , data=df , ax=ax[0] , color='teal')
ax[0].title.set_text('height_cm')
sns.histplot(x='weight_kg' , data=df , ax=ax[1])
ax[1].title.set_text('weight_kg')

kg_mean =df.weight_kg.mean()
cm_mean =df.height_cm.mean()
print(f'average score in height_cm is: {cm_mean}')
print(f'average score in weight_kg is: {cm_mean}')
```
