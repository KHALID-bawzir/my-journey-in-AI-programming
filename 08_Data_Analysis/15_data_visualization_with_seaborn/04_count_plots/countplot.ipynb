```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_dataset = sns.load_dataset("tips")
df_dataset
```

```python
sns.countplot(x='day' ,data=df_dataset ,hue='day' , order=df_dataset['day'].value_counts().index )
plt.show()
```

```python
sns.countplot(x='day' ,data=df_dataset ,hue='sex' , order=df_dataset['day'].value_counts().index)
plt.show()
```

```python
ax =sns.countplot(x='day' ,data=df_dataset ,hue='day' , order=df_dataset['day'].value_counts(ascending=True).index )
for label in ax.containers:
    ax.bar_label(label)

plt.show()
```
