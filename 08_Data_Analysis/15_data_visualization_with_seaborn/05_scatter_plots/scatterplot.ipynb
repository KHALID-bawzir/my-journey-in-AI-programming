```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('2016.csv')
df.head(3)
```

```python
sns.scatterplot(data=df , y='Happiness Score' , x='Economy (GDP per Capita)')
```

```python
fig = plt.figure(figsize=(15,7))
sns.scatterplot(data=df , y='Happiness Score' , x='Economy (GDP per Capita)' , hue='Region' ,palette="pastel")
```

```python
fac = sns.FacetGrid(df ,col='Region' , hue='Region', palette="husl",col_wrap=5)
fac.map(sns.scatterplot , 'Economy (GDP per Capita)' ,'Happiness Score' )
fac.add_legend()
```

```python
fig = plt.figure(figsize=(15,7))
sns.scatterplot(data=df , y='Happiness Score' , x='Economy (GDP per Capita)' ,size='Region')
```
