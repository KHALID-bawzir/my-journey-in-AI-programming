```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('euro-daily-hist_1999_2022.csv')
df
```

```python
df = df.iloc[:, [0,1,4,-2]]
df.columns = ['Date' , 'Australian dollar' , 'Canadian dollar' , 'US dollar ']
df
```

```python
df = pd.melt(df , id_vars='Date' , value_vars=['Australian dollar',	'Canadian dollar','US dollar '], value_name='Euro rate' , var_name='Currency')
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date']>='2022-12-1'].reset_index(drop=True)
df['Euro rate'] = pd.to_numeric(df['Euro rate'])
print(f'Currencies:{df.Currency.unique()}\n')
print(df)
```

```python
us =df[df['Currency']=='US dollar '].reset_index(drop=True)
sns.lineplot(x='Date',y='Euro rate',data=us )
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=us , linestyle='dotted' ,color='green' ,linewidth=4).set(title='Euro_US')
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=us , marker = '*' , markerfacecolor='green', markersize=20).set(title='Euro_US')
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=df , hue='Currency',marker = '*' , markerfacecolor='green').set(title='Euro_US')
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=df , style='Currency').set(title='Euro_US' , xlabel='X' , ylabel='Y')
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=df , hue='Currency' , palette=['red','blue','green']).set(title='Euro_US' , xlabel='X' , ylabel='Y')
```

```python
fig = plt.subplots(figsize=(20,5))
sns.lineplot(x='Date',y='Euro rate',data=df , size='Currency' ,sizes=[5,15,10]).set(title='Euro_US' , xlabel='X' , ylabel='Y')
```
