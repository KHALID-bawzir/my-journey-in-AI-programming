```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_dataset = sns.load_dataset("titanic")
df_dataset
```

```python
fig = plt.subplots(figsize=(8,5))
sns.barplot(x='who' ,y='survived' ,data=df_dataset)
```

```python
fig = plt.subplots(figsize=(8,5))
sns.barplot(x='who' ,y='survived' ,data=df_dataset ,palette='hls' ,  hue='who' ,legend=False)
plt.show()
```

```python
sns.barplot(x='who' ,y='survived' ,data=df_dataset ,palette='hls' ,  hue='class')
plt.show()
```

```python
sns.barplot(x='who' ,y='survived' ,data=df_dataset ,palette='hls' , errorbar=None,  hue='class')
plt.show()
```

```python
sns.barplot(x='who' ,y='survived' ,data=df_dataset ,palette='hls' , errorbar=None, order=['man' ,'child' , 'woman'], hue='class')
plt.show()
```

```python
sns.barplot(x='who' ,y='survived' ,data=df_dataset ,palette='hls' , errorbar=None, order=['man' , 'woman'], hue='class')
plt.show()
```
