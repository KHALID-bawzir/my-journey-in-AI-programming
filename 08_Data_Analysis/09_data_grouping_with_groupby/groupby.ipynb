```python
import pandas as pd

df = pd.read_csv('Ecommerce.csv')
df.info()
```

```python
df.head()
```

```python
gr = df.groupby('Amount')
gr.groups
```

```python
gr.get_group(2000)
```

```python
gr = df.groupby(['Country' , 'Gender' , 'Amount'])
for name , gro  in gr:
    print(name)
    print(gro)
```

```python
ss = df.groupby('Gender')['Id'].count()
ss
```

```python
gs = df.groupby(['Country' , 'Gender' ])['Id'].count()
gs
```

```python
ga = df.groupby('Country')['Amount'].mean()
ga
```
