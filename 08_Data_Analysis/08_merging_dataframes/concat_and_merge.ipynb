```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'column_a' : [1, 2, 3, 4],
    'column_b': ['a', 'b', 'c', 'd'],
    'column_c': [True, True, False, True]
})
df1
```

```python
df2 = pd.DataFrame({
    'column_a' : [1, 2, 8, 5],
    'column_b': ['a', 'b', 't', 'f'],
    'column_c': [False, False, False, True]
})
df2
```

```python
df = pd.concat([df1,df2],keys=['df1' , 'df2'] )
df
```

```python
df = pd.concat([df1,df2] , ignore_index=True , axis=1)
df
```

```python
df = pd.concat([df1,df2] ,  axis=1)
df
```

merge
-

```python
df1 = pd.DataFrame({
    'column_a' : [1, 2, 3, 4],
    'column_b': ['a', 'b', 'c', 'd'],
    'column_c': [True, True, False, True]
})
df1
```

```python
df2 = pd.DataFrame({
    'column_a' : [1, 2, 8, 5],
    'column_b': ['a', 'b', 't', 'f'],
    'column_c': [False, False, False, True]
})
df2
```

```python
dfm = pd.merge(df1,df2 , on='column_a')
dfm
```

```python
dfm = pd.merge(df1,df2 , on=['column_a','column_b'])
dfm
```

```python
df1.rename(columns={'column_a':'new'} ,inplace=True)
df1
```

```python
dfm = pd.merge(df1,df2 , left_on='new' , right_on='column_a')
dfm
```
