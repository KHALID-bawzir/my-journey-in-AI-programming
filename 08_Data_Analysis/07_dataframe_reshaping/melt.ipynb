```python
import pandas as pd
```

```python
dff = pd.DataFrame({
    'name':['khalid','ahmed','fwaz','seed'],
    'age':[15,34,23,55],
    'test':['A+' , 'C-' , 'B+' , 'D-']
})
dff
```

```python
dff1 = dff.melt(id_vars=['name'])
dff1
```

```python
dff2 = dff.melt(id_vars=['name'] , var_name='A' , value_name='B')
dff2
```

```python
dff2.sort_values(by='A')
```
