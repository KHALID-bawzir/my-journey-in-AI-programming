الدوران على الصفوف
---

```python
import pandas as pd
df = pd.read_csv('Students.csv' , sep=';')
df.head()
```

```python
for index , row in df.loc[1:5].iterrows():
    print(index, row[['gender' ,'reading score']])#لا يمكن تعديل عدد الصفوف التي تريد اخذها هنا بل فوق
```

```python
for row in df[1:6].itertuples():
    print(row)
    print("========================================================" ,row.gender)
```
