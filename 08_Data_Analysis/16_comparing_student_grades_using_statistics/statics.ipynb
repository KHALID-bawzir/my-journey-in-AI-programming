```python
import pandas as pd

df = pd.read_csv('Students_exams.csv')

df.info()
```

```python
df.columns = [col.lower().replace(' ','_') for col in df.columns]
print(df.columns)
```

```python
statics = pd.DataFrame(columns=['statics' ,'math' , 'reading', 'writing'])
statics
```

```python
x = df['math_score'].count()
y = df['reading_score'].count()
c = df['writing_score'].count()

rows = {'statics' : 'count' , 'math' :x , 'reading' :y , 'writing':c}
d_rows= pd.DataFrame.from_records([rows])
statics = pd.concat([statics ,d_rows ] , ignore_index = True)
```

```python
x = df['math_score'].mean()
y = df['reading_score'].mean()
c = df['writing_score'].mean()

rows = {'statics' : 'mean' , 'math' :x , 'reading' :y , 'writing':c}
d_rows= pd.DataFrame.from_records([rows])
statics = pd.concat([statics ,d_rows ] , ignore_index = True)
```
