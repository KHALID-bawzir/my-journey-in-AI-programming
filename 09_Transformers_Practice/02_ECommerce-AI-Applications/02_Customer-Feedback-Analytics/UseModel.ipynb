```python
my_folder = '/content/drive/MyDrive/TransformesCod/02/Customer_Feedback_An'

path_model = '/content/drive/MyDrive/TransformesCod/02/Customer_Feedback_An/roberta_base_fine_tuned'


from transformers import RobertaTokenizer, RobertaForSequenceClassification
import torch

tokenizer = RobertaTokenizer.from_pretrained(path_model)
model = RobertaForSequenceClassification.from_pretrained(path_model)

def classify_class(model , tokenizer , text):
    tok_inputs = tokenizer(text, padding=True, truncation=True, return_tensors="pt",max_length = 512)
    outputs = model(**tok_inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits).item()
    if predicted_class == 0:
        return "negative"
    elif predicted_class == 1:
        return "neutral"
    else:
        return "positive"
```

```python
t1 = 'are you ok'
c1 = classify_class(model , tokenizer , t1)
print(c1)
```

```python
text1 = "It's ok"
c1 = classify_class(model , tokenizer , text1)
print(c1)
text2 = "Awesome"
c2 = classify_class(model, tokenizer, text2)
print(c2)
text3 = "Dissatisfied Not worth the money"
c3 = classify_class(model, tokenizer, text3)
print(c3)
```

```python

examples = [
    "The product is amazing and very high quality",       # إيجابي (Positive)
    "I hated this experience, the service was terrible",   # سلبي (Negative)
    "It is okay, not the best but works fine",            # معتدل (Neutral)
    "Fast delivery and excellent packaging",              # إيجابي (Positive)
    "Waste of money, I want a refund immediately",        # سلبي (Negative)
    "The size is a bit small, but the color is nice",     # معتدل (Neutral)
    "I will definitely buy from this store again",        # إيجابي (Positive)
    "The app keeps crashing every time I open it",        # سلبي (Negative)
    "It arrived on time as expected",                     # معتدل (Neutral)
    "Totally disappointed with the performance",          # سلبي (Negative)
]

for text in examples:
    result = classify_class(model, tokenizer, text)
    print(f"Text: {text} | Prediction: {result}")
```

```python
negative_examples = [
    "This is the worst purchase I have ever made",        # الأسوأ على الإطلاق
    "Horrible quality, it broke after the first use",     # جودة مريعة وانكسر فوراً
    "I am extremely disappointed and will never return",  # خيبة أمل شديدة
    "The customer support was rude and unhelpful",        # سوء تعامل من الدعم
    "Do not buy this, it is a complete scam",             # تحذير من عملية احتيال
    "Poor performance, it does not work as advertised",    # أداء ضعيف ومخالف للإعلان
    "It is a total waste of time and money",              # ضياع للوقت والمال
    "I am very angry with this slow service",             # غضب من بطء الخدمة
    "The item arrived damaged and the box was open",      # وصول المنتج تالفاً
    "I regret buying this product, very bad experience"   # ندم على الشراء
]

for text in negative_examples:
    result = classify_class(model, tokenizer, text)
    print(f"Text: {text} | Prediction: {result}")
```
