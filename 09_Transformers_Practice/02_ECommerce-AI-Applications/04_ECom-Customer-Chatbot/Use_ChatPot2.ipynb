```python
from torch import trunc
model_path = '/content/drive/MyDrive/TransformesCod/02/ChatBot/model'

from transformers import AutoTokenizer, AutoModelForCausalLM
model = AutoModelForCausalLM.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)

def answer_question(question, model, tokenizer, max_length=40):
  tokenized_input = tokenizer(question, return_tensors='pt' , truncation=True, max_length=max_length , padding =True)
  input_ids = tokenized_input['input_ids']
  attention_mask = tokenized_input['attention_mask']
  out_put = model.generate(input_ids, attention_mask=attention_mask,max_length = max_length , num_return_sequences=1 ,pad_token_id = tokenizer.eos_token_id)

  response = tokenizer.decode(out_put[0], skip_special_tokens=True)
  return response
```

```python
question = "any physical store?"
answer = answer_question(question, model, tokenizer)
print(f"Q/A: {answer}\n")
```

```python
t = ' offer expedited shipping?'
answer = answer_question(t, model, tokenizer)
print(f"Q/A: {answer}\n")
```
