```python
import pandas as pd
import os

my_folder = '/content/drive/MyDrive/TransformesCod/02/ProductRecommendations/'

file_csv = os.path.join(my_folder, 'flipkart_com-ecommerce_sample.csv')
df = pd.read_csv(file_csv)
df = df[['product_name']]
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
import numpy as np
embeddings_path = my_folder + 'embeddings.npy'
sentence_embeddings = np.load(embeddings_path)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import re
def get_product(user_input , sentence_embeddings , model , df , num_recommendations = 10):
  user_input = re.sub(r'[^A-Za-z0-9]+', ' ', user_input)
  user_input = user_input.lower()
  user_embedding = model.encode([user_input])
  similarities = cosine_similarity(user_embedding, sentence_embeddings)
  similarities=similarities[0]
  indices_asc = similarities.argsort()
  indices_asc = indices_asc[::-1]
  top_indices = indices_asc[:num_recommendations]
  re_pr = df.iloc[top_indices]['product_name']
  return re_pr
```

```python
in_p = "bluetooth wireless headphones"

res = get_product(in_p , sentence_embeddings , model , df )
print("Recommendations for \"", in_p, "\" are: ")
print(res)
```
